```
s=match.groupby(['winner'])['win_by_runs'].sum()
s.sort_values(ascending=False,inplace=True)
fig,ax=plt.subplots(figsize=(20,10))
plt.title("Winner based on runs") #Winner based on max runs
s.plot.bar()
```
