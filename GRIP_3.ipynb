```
x=data['State']
```

```
y=data['Sales']
```

```
#To find out which state has maximum Sales
s=data.groupby(['State'])['Sales'].sum()
s.sort_values(ascending=False,inplace=True)
fig,ax=plt.subplots(figsize=(20,10))
plt.title('Statewise Sales')
s.plot.bar()
```

```
plt.figure(figsize=(20,20))
data['Sub-Category'].value_counts().plot.pie(autopct='%1.1f%%')
```

```
plt.figure(figsize=(15,12))
sns.barplot(x='Sub-Category',y='Profit',data=data,palette='Spectral')
```

```
#To find which state has Maximum profit
s=data.groupby(['State'])['Profit'].sum()
s.sort_values(ascending=False,inplace=True)
fig,ax=plt.subplots(figsize=(20,10))
plt.title('Statewise Profit')
s.plot.bar()
```

```
#Profit based on Sub-Category
s=data.groupby(['Sub-Category'])['Profit'].sum()
s.sort_values(ascending=False,inplace=True)
fig,ax=plt.subplots(figsize=(20,10))
plt.title('Profit based on sub-category')
s.plot.bar()
```

```
a=data['Category']
```

```
#Impact of Sales on the Basis of Category
plt.bar(a,y,color='r')
plt.xlabel('X Axis')
plt.ylabel('Y_label')
plt.title('Sales with respect to Category')
```

```
#Profit and Sales based on Sub-Category
s=data.groupby(['Sub-Category'])['Profit','Sales'].agg(['sum'])
s.plot.bar(figsize=(15,10),rot=0)
plt.title('Profit and Sales based on Sub-Category')
```

```
#Profit and sales based on Category
s=data.groupby(['Category'])['Profit','Sales'].agg(['sum'])
s.plot.bar(figsize=(15,10),rot=0)
plt.title('Profit and Sales based on Category')
```
